# Preprocessing Dataset

## 1. Lowercase

```python
data_PHI['Perkara'] = data_PHI['Perkara'].str.lower() #ubah hruf besar ke kecil
data_PHI
```

## 2. Remove Single Char

```python
def remove_single_char (text):
  text = re.sub(r"\b[a-zA-Z]\b", "",text)
  return text

data_PHI['Perkara'] = data_PHI['Perkara'].apply(remove_single_char)
print(data_PHI['Perkara'])
```

## Remove Numbers

```python
def remove_numbers (text):
  text = re.sub(r'[0-9]+', "",text)
  return text

data_PHI['Perkara'] = data_PHI['Perkara'].apply(remove_numbers)
print(data_PHI['Perkara'])
```

## Remove White Space

```python
def remove_whitespace(df):
  return df.strip()

data_PHI['Perkara'] = data_PHI['Perkara'].apply(remove_whitespace)
print(data_PHI['Perkara'])
```

## Tokenize

```python
import nltk
nltk.download('punkt')
```

```python
def tokenized(data):
  return data.apply(nltk.word_tokenize)
```

```python
data_PHI['Perkara']=tokenized(data_PHI['Perkara'])
data_PHI.head()
```

## Steamming

```python
!pip install Sastrawi
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
def stemming(data):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()

  return [stemmer.stem(perkara) for perkara in data]
```

```python
data_PHI['Perkara'] = data_PHI['Perkara'].apply(stemming)
data_PHI.head()
```

## STop WOrds Removal

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
def removeStopWords(data):
  stop_words = set(stopwords.words('indonesian'))
  return data.apply(lambda x: [item for item in x if item not in stop_words])
```

```python
data_PHI['Perkara']=removeStopWords(data_PHI['Perkara'])
data_PHI.head()
```

## Save Preprocessing Result

```python
# combine cleaned text after removing stop words into one string
data_PHI['Perkara-Final'] = data_PHI['Perkara'].apply(lambda x: ' '.join(x))
data_PHI.head()
```

```python
data1 = data_PHI['Perkara-Final']
data2 = data_PHI['Label']
dataset = pd.concat([data1, data2], axis=1)
dataset.head()
```
